### Problem 1
Q1. Create a CSV file containing product names and prices. Read the data using
Pandas. Use NumPy to calculate the average price. Write a function to identify
products priced above the average and display them.
Concepts Used:


 Pandas DataFrame


 NumPy


 Functions


 Control Statements

File creation

```python
import pandas as pd

data = {
    "Product": ["Laptop", "Mouse", "Keyboard", "Monitor", "Headphones", "Printer"],
    "Price": [60000, 800, 1500, 12000, 2500, 9000]
}

df = pd.DataFrame(data)

# Create CSV file
df.to_csv("products.csv", index=False)

print("products.csv file created successfully.")
```

Solution

```python
import pandas as pd
import numpy as np

# Read the CSV file
df = pd.read_csv("products.csv")

# Calculate average price using NumPy
average_price = np.mean(df["Price"])

print("Products Data:")
print(df)

print("\nAverage Price =", average_price)

# Function to display products priced above average
def above_average_products(data, avg):
    print("\nProducts priced above the average:")

    found = False
    for i in range(len(data)):
        if data.loc[i, "Price"] > avg:
            print(data.loc[i, "Product"], "-", data.loc[i, "Price"])
            found = True

    if not found:
        print("No products found.")

# Function call
above_average_products(df, average_price)
```

# **Problem 2**

Q2. Read daily temperature data from a CSV file using Pandas. Use NumPy to
calculate the highest, lowest, and average temperatures. Write functions to classify
each day as:

 Cold (&lt;20°C)

 Moderate (20–30°C)

 Hot (&gt;30°C)

Display the category for each day.
Concepts Used:

 Pandas

 NumPy

 Functions

 if-elif-else

```python
import pandas as pd

data = {
    "Day": ["Monday", "Tuesday", "Wednesday", "Thursday", "Friday", "Saturday", "Sunday"],
    "Temperature": [18, 25, 32, 29, 35, 16, 27]
}

df = pd.DataFrame(data)

# Create CSV file
df.to_csv("temperature.csv", index=False)

print("temperature.csv file created successfully.")
```

Solution

```python
import pandas as pd
import numpy as np

# Read CSV file
df = pd.read_csv("temperature.csv")

# Calculate highest, lowest, and average temperature
highest = np.max(df["Temperature"])
lowest = np.min(df["Temperature"])
average = np.mean(df["Temperature"])

print("Highest Temperature:", highest)
print("Lowest Temperature:", lowest)
print("Average Temperature:", average)

# Function to classify temperature
def classify(temp):
    if temp < 20:
        return "Cold"
    elif temp <= 30:
        return "Moderate"
    else:
        return "Hot"

print("\nTemperature Category:")

# Display category for each day
for i in range(len(df)):
    day = df.loc[i, "Day"]
    temp = df.loc[i, "Temperature"]
    print(day, "-", temp, "°C -", classify(temp))
```
