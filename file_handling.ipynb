```python
file4 = open(path + "/student.txt", "r")

pass_count = 0
fail_count = 0

passed_students = []
failed_students = []

for i in file4:
    data = i.strip().split(",")

    name = data[0]
    marks = int(data[1])

    if marks > 45:
        passed_students.append(name)
        pass_count += 1
    else:
        failed_students.append(name)
        fail_count += 1

file4.close()

print("Passed Students:")
for student in passed_students:
    print(student)

print("\nFailed Students:")
for student in failed_students:
    print(student)

print("\nTotal Passed:", pass_count)
print("Total Failed:", fail_count)
```

```python
file = open("salary.txt", "w")

file.write("Rohan 45000\n")
file.write("Aman 60000\n")
file.write("Priya 75000\n")
file.write("Neha 48000\n")
file.write("Rahul 55000\n")
file.write("Simran 65000\n")
file.write("Karan 52000\n")
file.write("Anjali 40000\n")

file.close()

print("salary.txt created successfully.")
```

```python
count = 0

file = open("salary.txt", "r")

print("Employees with salary greater than ₹50,000:")

for line in file:
    data = line.split()

    name = data[0]
    salary = int(data[1])

    if salary > 50000:
        print(name)
        count += 1

file.close()

print("Total Employees:", count)
```

```python
file = open("attendance.txt", "w")

file.write("Rohan Present\n")
file.write("Aman Absent\n")
file.write("Priya Present\n")
file.write("Neha Absent\n")
file.write("Rahul Present\n")
file.write("Simran Present\n")
file.write("Karan Absent\n")
file.write("Anjali Present\n")

file.close()

print("attendance.txt created successfully.")
```

```python
present = 0
absent = 0

file = open("attendance.txt", "r")

print("Absent Students:")

for line in file:
    data = line.split()

    name = data[0]
    status = data[1]

    if status == "Absent":
        print(name)
        absent += 1
    else:
        present += 1

file.close()

print("Total Present Students:", present)
print("Total Absent Students:", absent)
```
