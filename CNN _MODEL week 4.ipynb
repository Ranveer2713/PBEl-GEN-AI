```python
# libraries for discriminator
from keras.models import Sequential
from keras.layers import Dense,Dropout,Flatten,Reshape,LeakyReLU
from keras.layers import Conv2D
from keras.optimizers import Adam
from keras.utils import plot_model
import numpy as np
```

```python
def define_discriminator(in_shape =(32,32,3)):
  model=sequential()
  model.add(Convo2D(64,(3,3),padding ='same',input_shape= in_shape))
  model.add(LeakyRelu(alpha=0.2))

  model.add(Convo2D(128,(3,3),strides =(2,2),padding ='same))
  model.add(LeakyRelu(alpha=0.2))

  model.add(Convo2D(128,(3,3),strides =(2,2),padding ='same))
  model.add(LeakyRelu(alpha=0.2))

  model.add(Convo2D(256,(3,3),strides =(2,2),padding ='same))
  model.add(LeakyRelu(alpha=0.2))

  model.add(Flatten())
  model.add(Dropout(0.4))
  model.add(Dense(1,activation ='sigmoid'))

  opt = Adam(lr= 0.0002,beta_1=0.Adam)
  model.compile(loss='binary_crossentropy',optimizer=opt,metrics=['accuracy'])
   return model




```
